```python
import re

# UDlexPT - léxico geral
import sys
sys.path.append('/content/drive/MyDrive/IC/Recursos/PortiLexicon-UD')
from UDlexPT import UDlexPT
UDlexPT = UDlexPT()

# pontuações (que serão removidas dos tweets)
import string
pontuacoes = string.punctuation + '’‘“”ºª°…—–«»¯´'

# Enelvo - normalizador
sys.path.append('/content/drive/MyDrive/IC/Recursos/')
from enelvo import normaliser
norm = normaliser.Normaliser(fc_list='/content/drive/MyDrive/IC/Recursos/force-list.txt',ig_list='/content/drive/MyDrive/IC/Recursos/ignore-list.txt')

# spaCy - biblioteca para PLN
import spacy
nlp = spacy.load('pt_core_news_lg')  # carregando o modelo em português

# pyspellchecker - corretor ortográfico
from spellchecker import SpellChecker
spellchecker = SpellChecker(language='pt')  # inicializando o corretor para o idioma português

import unicodedata

def remover_acentos(palavra):
    return ''.join(c for c in unicodedata.normalize('NFD', palavra) if unicodedata.category(c) != 'Mn')

def eh_risada(word):
    return bool(re.fullmatch(r'(ha)+|k+', word, re.IGNORECASE))

def eh_link(word):
    return bool(re.match(r'(http|www)\.|\.com|\.br|\.org|@\w+', word))

def corrigir_hifen_palavras(palavra):
    if re.match(r'^(ex|vice)[a-záàâãéèêíïóôõöúçñ]+$', palavra, re.IGNORECASE):
        return re.sub(r'^(ex|vice)', r'\1-', palavra)
    if re.match(r'^(pós|pré|pró)[a-záàâãéèêíïóôõöúçñ]+$', palavra, re.IGNORECASE):
        return re.sub(r'^(pós|pré|pró)', r'\1-', palavra)
    return palavra


def eh_termo_tecnico(word, entidades_per_gpe, pos):
    return (not UDlexPT.exists(word) and  # a palavra não está no léxico geral
            word not in entidades_per_gpe and  # não é pessoa ou local
            word not in df_paises['nome'].values and
            word not in df_estados['nome'].values and
            word not in df_municipios['nome'].values and
            remover_acentos(word.lower()) not in df_nomes_fem['nome'].values and  # não é nome feminino
            remover_acentos(word.lower()) not in df_nomes_masc['nome'].values and  # não é nome masculino
            not eh_risada(word) and
            not eh_link(word) and
            pos not in {'ADP', 'SCONJ', 'PRON', 'DET'} and  # não é preposição, conjunção subordinativa, pronome ou determinante
            word.isalnum() and  # é alfanumérica
            not word.isnumeric() and  # não é numérica
            len(word) > 1)  # tem mais de um caractere


termos_tecnicos = set()
df = pd.read_excel(MEU_DRIVE + '/Recursos/Amostras/Cópia de tweets-2022_10_10-query_bolsonaro.xlsx')

for tweet in df['text'][2000:]:
    norm_sentence = norm.normalise(tweet)  # normalizando o tweet
    doc = nlp(norm_sentence)
    entidades_per_gpe = {ent.text for ent in doc.ents if ent.label_ in {'PER', 'GPE', 'LOC'}}
    for token in doc:
        word = token.text.translate(str.maketrans('', '', pontuacoes))  # removendo pontuações do token
        word = corrigir_hifen_palavras(word)
        if not UDlexPT.exists(word):
            correcao = spellchecker.correction(word)
            if correcao:
                word = correcao  # corrigindo a ortografia da palavra
        # se a palavra for um termo técnico, adiciona ao conjunto
        if eh_termo_tecnico(word, entidades_per_gpe, token.pos_):
            termos_tecnicos.add(word)
termos_tecnicos = sorted(termos_tecnicos)

# Termos técnicos agrupados por letra inicial
letra_atual = ''
termos_para_letra = []
for termo in termos_tecnicos:
    letra_inicial = termo[0].upper()
    if letra_inicial != letra_atual:
        if termos_para_letra:
            print(f'"{letra_atual}": {termos_para_letra}')
        letra_atual = letra_inicial
        termos_para_letra = [termo]
    else:
        termos_para_letra.append(termo)
if termos_para_letra:
    print(f'"{letra_atual}": {termos_para_letra}')
```
